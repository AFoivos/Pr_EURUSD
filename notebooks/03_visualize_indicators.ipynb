```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/EURUSD_with_indicators.csv", parse_dates=["datetime"])
df.set_index("datetime", inplace=True)
```

```python
plot_df = df.iloc[-200:].copy()
```

```python
# Price + EMA + Bollinger Bands
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(plot_df.index, plot_df["<CLOSE>"], label="Close", color="black")
ax.plot(plot_df.index, plot_df["ema_10"], label="EMA 10", color="blue")
ax.plot(plot_df.index, plot_df["ema_20"], label="EMA 20", color="green")
ax.plot(plot_df.index, plot_df["ema_50"], label="EMA 50", color="orange")
ax.plot(plot_df.index, plot_df["bb_upper"], label="BB Upper", linestyle="--", color="red")
ax.plot(plot_df.index, plot_df["bb_lower"], label="BB Lower", linestyle="--", color="red")
ax.set_title("Price + EMA + Bollinger Bands")
ax.legend()
plt.grid(True)
plt.show()
```

```python
# Price + EMA 50/100/200 
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(plot_df.index, plot_df["<CLOSE>"], label="Close", color="black")
ax.plot(plot_df.index, plot_df["ema_50"], label="EMA 50", color="orange")
ax.plot(plot_df.index, plot_df["ema_100"], label="EMA 100", color="blue")
ax.plot(plot_df.index, plot_df["ema_200"], label="EMA 200", color="green")
ax.set_title(" Price + EMA 50/100/200")
ax.legend()
plt.grid(True)
plt.show()
```

```python
# RSI 
fig, ax = plt.subplots(figsize=(20, 5))
plot_df['rsi'].plot(ax=ax, color='purple')
ax.axhline(70, color='red', linestyle='--')
ax.axhline(30, color='green', linestyle='--')
ax.set_title("RSI")
ax.set_ylim(0, 100)
plt.grid(True)
plt.show()
```

```python
# MACD 
fig, ax = plt.subplots(figsize=(20, 5))
plot_df['macd'].plot(ax=ax, label='MACD', color='blue')
plot_df['macd_signal'].plot(ax=ax, label='Signal', color='red')
ax.bar(plot_df.index, plot_df['macd_diff'], label='Histogram', color='gray', alpha=0.4)
ax.legend()
ax.set_title("MACD")
plt.grid(True)
plt.show()
```

```python
# Stochastic Oscillator
fig, ax = plt.subplots(figsize=(20, 5))
plot_df['stoch_k'].plot(ax=ax, label='%K', color='blue')
plot_df['stoch_d'].plot(ax=ax, label='%D', color='orange')
ax.axhline(80, color='red', linestyle='--')
ax.axhline(20, color='green', linestyle='--')
ax.set_title("Stochastic Oscillator")
ax.set_ylim(0, 100)
ax.legend()
plt.grid(True)
plt.show()
```

```python
# ATR Plot
fig, ax = plt.subplots(figsize=(20, 5))
plot_df['atr'].plot(ax=ax, color='brown')
ax.set_title("ATR (Average True Range)")
plt.grid(True)
plt.show()
```

```python
# ADX Plot 
fig, ax = plt.subplots(figsize=(20, 5))
plot_df['adx'].plot(ax=ax, label='ADX', color='black')
plot_df['adx_pos'].plot(ax=ax, label='+DI', color='green')
plot_df['adx_neg'].plot(ax=ax, label='-DI', color='red')
ax.set_title("ADX")
ax.legend()
plt.grid(True)
plt.show()
```
